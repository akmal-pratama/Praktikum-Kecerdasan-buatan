```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Membaca dataset student-por.csv ke dalam DataFrame df. Separator yang dipakai adalah ; karena dataset dari UCI Student Performance memang pakai pemisah titik koma.

```python
# Tentukan path ke file Anda
file_path = 'student-por.csv'

# Baca file CSV
try:
    df = pd.read_csv(file_path, sep=';')
    print("Dataset berhasil masuk ke dalam DataFrame 'df'.")
except FileNotFoundError:
    print(f"ERROR: File '{file_path}' tidak ditemukan.")
    print("file nya gak ke baca di Jupyter Notebook nya.")
```

```python
df.head()
```

### Menampilkan informasi umum dataset:
### Jumlah baris (record)
### Jumlah kolom (atribut)
### Jumlah atribut numerik

```python
print("### 3. Informasi Umum Dataset ###")
print(f"Jumlah Record          : {df.shape[0]}")
print(f"Jumlah Atribut         : {df.shape[1]}")
numeric_cols = df.select_dtypes(include=np.number)
print(f"Jumlah Atribut Angka   : {numeric_cols.shape[1]}")
print("\n" + "="*80 + "\n")
```

### Menampilkan statistik deskriptif untuk atribut numerik: rata-rata, standar deviasi, nilai minimum, kuartil (Q1, Q2, Q3), dan nilai maksimum.

```python
print("### 1. Statistik Deskriptif Atribut Angka ###")
descriptive_stats = numeric_cols.describe().transpose().rename(columns={
    'mean': 'Rata-rata', 'std': 'Standar Deviasi', 'min': 'Nilai Minimum',
    'max': 'Nilai Maksimum', '25%': 'Q1', '50%': 'Q2 (Median)', '75%': 'Q3'
})
print(descriptive_stats)
print("\n" + "="*80 + "\n")
```

### Menampilkan 10 record pertama khusus untuk kolom numerik agar lebih fokus melihat data angka seperti usia, absensi, nilai G1â€“G3, dll.

```python
print("### 2. 10 Record Pertama dari Atribut Angka ###")
print(numeric_cols.head(10))
print("\n" + "="*80 + "\n")
```

### Menampilkan jumlah label (frekuensi) untuk setiap atribut kategorikal (tipe object).

```python
print("### 4. Jumlah Label untuk Atribut Kategorikal (Object) ###")
object_cols = df.select_dtypes(include=['object'])
for col in object_cols.columns:
    print(f"\n----- Rincian Atribut: {col} -----")
    print(object_cols[col].value_counts())
print("\n" + "="*80 + "\n")
```

### Membuat boxplot nilai akhir (G3) untuk melihat distribusi nilai siswa:
### Median
### Rentang nilai mayoritas siswa
### Outlier (siswa dengan nilai sangat rendah/tinggi)

```python
print("### 5. Visualisasi Distribusi Nilai Akhir (G3) dengan Boxplot ###\n")
plt.figure(figsize=(12, 7))
sns.boxplot(x='G3', data=df, orient='h', palette='viridis')
plt.title('Distribusi Nilai Akhir Siswa (G3)', fontsize=16)
plt.xlabel('Nilai Akhir (Skala 0-20)', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

### Warna merah/biru menunjukkan kekuatan korelasi (positif/negatif).
### Terlihat bahwa G1, G2, dan G3 punya korelasi yang kuat â†’ artinya nilai ujian awal memengaruhi nilai akhir.

```python
print("### 6. Visualisasi Korelasi Antar Atribut Numerik dengan Heatmap ###\n")
plt.figure(figsize=(16, 12))
correlation_matrix = numeric_cols.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap Korelasi Antar Atribut Numerik', fontsize=18)
plt.show()
```

## Kesimpulan Analisis dan Visualisasi Data

1. **Informasi Dataset**  
   Dataset berisi data siswa dengan jumlah record cukup besar serta atribut numerik dan kategorikal yang beragam.  
   Atribut numerik mencakup nilai (G1, G2, G3) serta data lain seperti absensi dan usia, sedangkan atribut kategorikal mencakup informasi pribadi (misalnya sekolah, jenis kelamin, status keluarga, pekerjaan orang tua, dll).

2. **Statistik Deskriptif**  
   - Nilai rata-rata untuk atribut numerik menunjukkan rentang umum siswa.  
   - Terlihat variasi yang cukup besar pada nilai akhir (G3), dengan standar deviasi yang menunjukkan adanya perbedaan performa antar siswa.  
   - Sebagian atribut lain (misalnya absensi) memiliki distribusi miring, artinya ada kelompok kecil siswa dengan nilai ekstrim.

3. **Distribusi Nilai Akhir (G3)**  
   - Hasil boxplot menunjukkan sebagian besar nilai akhir siswa berada pada kisaran menengah.  
   - Ada beberapa siswa yang memperoleh nilai sangat rendah (outlier di bagian bawah).  
   - Nilai maksimum mencapai 20, yang menunjukkan ada siswa dengan performa sempurna.

4. **Korelasi Antar Atribut Numerik**  
   - Terdapat korelasi positif yang cukup kuat antara nilai G1, G2, dan G3, artinya performa siswa pada ujian/tes sebelumnya berhubungan erat dengan hasil akhir.  
   - Atribut lain seperti absensi, waktu belajar, atau konsumsi alkohol menunjukkan korelasi lebih lemah dengan nilai akhir.  
   - Secara umum, hasil akhir siswa lebih dipengaruhi oleh konsistensi nilai pada penilaian sebelumnya.

### Ringkasan
Dari analisis ini dapat disimpulkan bahwa:
- Nilai siswa sangat dipengaruhi oleh konsistensi performa sebelumnya (G1 dan G2).  
- Terdapat disparitas antar siswa, dengan sebagian kecil yang memperoleh nilai sangat rendah maupun sangat tinggi.  
- Faktor eksternal selain nilai ujian awal tampaknya tidak memberikan korelasi kuat terhadap hasil akhir, sehingga peningkatan belajar secara konsisten lebih menentukan keberhasilan.
