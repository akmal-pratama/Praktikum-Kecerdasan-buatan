```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('canadian_immegration_data.csv')
df.head()
```

```python
df.shape
```

```python
df['Country'].unique()
```

```python
df.set_index('Country', inplace=True)
```

```python
df.head()
```

## bar plot 

```python
df['Continent'].unique()
```

```python
jumlah_continent = df['Continent'].value_counts()
jumlah_continent
```

```python
continent_group = df.groupby('Continent').sum()
continent_group.head()
```

```python
continent_group['Total'].plot.bar()
```

```python
plt.bar(continent_group.index,
       continent_group['Total'])
plt.title('Total Imigran')
plt.xlabel('Benua (Continent)')
plt.ylabel('Total')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
sns.barplot(
    x=continent_group.index,
    y=continent_group['Total'],
    palette='pastel')
plt.xticks(rotation=45, ha='right')
```

**berdasarkan grafik di atasa dapat ditarik kesimpulan bahwa benua dengan jumlah imigran paling banyak adalah asia dengan total....**
